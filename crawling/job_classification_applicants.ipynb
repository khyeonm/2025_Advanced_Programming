```python
def classify_it_job(detail):
    detail = str(detail).lower().strip()

    # ✅ 서버 / 백엔드
    if any(k in detail for k in ['백엔드', 'backend', 'server', '업무원장개발', '서비스 개발', 'api']):
        return '서버·백엔드'

    # ✅ 프론트엔드 / 웹
    elif any(k in detail for k in ['프론트엔드', '퍼블리싱', 'ui', 'ux', '웹', 'frontend']):
        return '프론트엔드 / 웹'

    # ✅ IT 기획 / 컨설팅
    elif any(k in detail for k in ['기획', '컨설팅', '신사업', '전략기획', '전략']):
        return 'IT기획 / 컨설팅'

    # ✅ IT 운영 / 시스템
    elif any(k in detail for k in ['운영', '전산', 'infra', '인프라', '시스템', '아키텍처']):
        return 'IT운영 / 시스템'

    # ✅ 데이터 / AI
    elif any(k in detail for k in ['데이터', 'ai', '머신러닝', 'deep learning', 'r&d', '빅데이터', '금융공학']):
        return '데이터 / AI / R&D'

    # ✅ 정보보안
    elif any(k in detail for k in ['보안', 'security', '정보보호', 'cyber']):
        return '정보보안 / 보안엔지니어'

    # ✅ 소프트웨어 개발
    elif any(k in detail for k in ['sw', '소프트웨어', '응용', 'programming', '개발', '프로그래머', 'it programmer']):
        return '소프트웨어 개발'

    # ✅ QA / 테스트
    elif any(k in detail for k in ['qa', '품질', '검증', '테스트']):
        return 'QA / 테스트'

    # ✅ IT 영업
    elif any(k in detail for k in ['영업', 'sales']):
        return 'IT영업 / 기술영업'

    # ✅ 디지털 / 스마트IT
    elif any(k in detail for k in ['디지털', 'smart', '스마트', '애자일']):
        return '디지털 / 스마트IT'

    # ✅ 네트워크
    elif any(k in detail for k in ['network', 'it infrastructure']):
        return '네트워크 / 인프라'

    # ✅ 금융IT
    elif any(k in detail for k in ['금융it', '카드', 'cfo', '금융', '은행', 'it업무혁신']):
        return '금융 IT'

    # ✅ 연구 / 학술
    elif any(k in detail for k in ['석박사', '연구개발', '연구', '인재육성', 'academy']):
        return '연구개발 / 학술IT'

    # ✅ 전사 지원
    elif any(k in detail for k in ['전사', '사업부문', '지원', 'staff']):
        return '전사 IT 지원'

    # ✅ 기타 잡힘 방지용: IT 단어만 있는 경우
    elif detail in ['it', 'it분야', 'it 직무', 'it계열', 'it인재(3급)', 'it서비스', 'it관리', 'it cloud']:
        return 'IT 일반'

    else:
        return '기타 IT'


def classify_production_job(detail):
    detail = str(detail).lower().strip()
    
    # 생산관리
    if any(k in detail for k in ['생산관리', '생산운영', '생산기획', '생산팀', '생산직무', '생산판매', '출하']):
        return '생산관리'
    
    # 생산기술
    elif any(k in detail for k in ['생산기술', '생산 기술', '공정기술', '제조기술', '제조공정', '공정관리', '생산기술직', '생산 엔지니어']):
        return '생산기술'
    
    # 생산 일반
    elif any(k in detail for k in ['생산', '생산공정', '생산부문', '생산운전', '생산스탭', '생산 직무', '생산(공장)', '생산1팀']):
        return '생산 (일반)'
    
    # 제조 / 생산지원
    elif any(k in detail for k in ['생산지원', '제조', '제조부문', '제조전략', '제조솔루션', '제조지원', '제조원가', '구매', '소싱']):
        return '제조 / 생산지원'
    
    # 품질관리
    elif any(k in detail for k in ['품질', '품질관리', '품질보증', '생산품질', '품질생산', '품진생산']):
        return '품질관리'
    
    # 설비 / 유지보수
    elif any(k in detail for k in ['설비', '정비', '설비관리', '설비운영', '유지보수', 'esh', '전기공무', '기술관리']):
        return '설비 / 유지보수'
    
    # 노무 / 현장
    elif any(k in detail for k in ['노무', '기능직', '현장', '직공', '생산직', '오퍼레이터', '기술직(생산)']):
        return '노무 / 현장'
    
    # R&D
    elif any(k in detail for k in ['r&d', '연구', '연구개발', '기술개발', '연구소']):
        return 'R&D / 연구개발'
    
    # 배터리 / 전지 / 반도체
    elif any(k in detail for k in ['배터리', '전지', '이차전지', '공정생산', '반도체']):
        return '배터리 / 전지'
    
    # 화학 / 바이오 / 소재
    elif any(k in detail for k in ['bio', '생명과학', '정제', '발효', '제약', '화학', '석유', '합성수지', '소재', '기초소재', '첨단소재', '기판소재', '재료', '재료사업부', '산업자재']):
        return '화학 / 바이오 생산'
    
    # 자동차 / 기계
    elif any(k in detail for k in ['자동차', '차량', '기계', '엔진', '파워트레인', '차체']):
        return '자동차 / 기계 생산'

    # 기타 엔지니어링 부서 (DS, DX, LED, H&A 등)
    elif any(k in detail for k in ['dx부문', 'ds생산', 'led', 'ce', 'h&a', 'p']):
        return '전자 / 디지털 생산'

    # 이상치 / 불분명
    elif detail in ['#name?', 'nan', '']:
        return '기타'
    
    else:
        return '기타'
    


def classify_detail_job(detail):
    detail = str(detail).lower().strip()

    # 🎯 연구개발
    if any(k in detail for k in ['재료', '소재', 'compound', 'mim', '원료']):
        return '연구개발 - 재료/소재'
    elif any(k in detail for k in ['전기', '전자', '전장', '아날로그']):
        return '연구개발 - 전기/전자'
    elif any(k in detail for k in ['기계', '전동', '액추에이터']) and '설계' not in detail:
        return '연구개발 - 기계'
    elif any(k in detail for k in ['bio', '생명과학', '제약', '정제', '발효']):
        return '연구개발 - 화학/바이오'
    elif any(k in detail for k in ['자율주행', '차량', '전장sw', '제어']):
        return '연구개발 - 차량/자율주행'
    elif any(k in detail for k in ['반도체', 'system lsi', 'lsi', '파운드리']):
        return '연구개발 - 반도체'
    elif any(k in detail for k in ['ai', 'multimodal', 'llm']):
        return '연구개발 - AI/데이터'
    elif any(k in detail for k in ['공정개발', '공정기술', '공정연구']):
        return '연구개발 - 공정'

    # 🎯 설계
    elif any(k in detail for k in ['기구설계', '기계설계', '구조설계']):
        return '설계 - 기계/기구'
    elif any(k in detail for k in ['회로설계', 'soc']):
        return '설계 - 회로'
    elif any(k in detail for k in ['전기설계', '제어설계']):
        return '설계 - 전기'
    elif any(k in detail for k in ['건축', '조경', '인테리어', '플랜트', '공사관리', '공조']):
        return '설계 - 건축/플랜트'
    elif any(k in detail for k in ['system', '솔루션', 'solution설계', '신호 및 시스템']):
        return '설계 - 시스템'

    # 🎯 소프트웨어 개발
    elif '백엔드' in detail or 'backend' in detail:
        return '소프트웨어 개발 - 백엔드'
    elif '프론트엔드' in detail or 'frontend' in detail:
        return '소프트웨어 개발 - 프론트'
    elif '웹개발' in detail or 'web' in detail:
        return '소프트웨어 개발 - 웹'
    elif any(k in detail for k in ['fw', 'platform', 'ssd']):
        return '소프트웨어 개발 - 플랫폼'

    # 🎯 IT 시스템 개발
    elif any(k in detail for k in ['시스템 운영', '서비스 개발', 'ict개발', 'it개발', 'it 개발', '서비스개발', 'plm']):
        return '시스템개발/IT'
    elif any(k in detail for k in ['금융 it', '금융it']):
        return '금융IT'

    # 🎯 제품개발
    elif any(k in detail for k in ['제품개발', '상품개발', '양산개발', '패키지개발']):
        return '제품개발'

    # 🎯 UX / 콘텐츠 / 인턴
    elif any(k in detail for k in ['ux', 'ui', '콘텐츠', '인턴', '유아동']):
        return 'UX/UI/콘텐츠'

    # 🎯 전략 / 사업개발
    elif any(k in detail for k in ['사업개발', 'global', 'trading', '사업팀', '미래전략']):
        return '전략/사업개발'

    # 🎯 기술 연구소 / 부서
    elif any(k in detail for k in ['혁신센터', '연구소', 'dx부문', 'ds부문', 'cto']):
        return '기술연구소/센터'

    # 🎯 생산 / 공정
    elif any(k in detail for k in ['생산관리', '생산기술']):
        return '생산관리'
    elif detail == '생산':
        return '생산 (일반)'

    # 🎯 조사 / 학술
    elif any(k in detail for k in ['학술개발', '조사']):
        return '학술/조사'

    # 🎯 HRD
    elif any(k in detail for k in ['hrd', '교육']):
        return 'HRD/교육'

    # 🎯 일반 키워드 (잡아주는 용)
    elif any(k in detail for k in ['연구개발', 'r&d', '연구분야', '연구개발팀']):
        return '연구개발 (일반)'
    elif detail == '설계':
        return '설계 (일반)'
    elif detail == '개발' or '개발자' in detail or 'windows 개발' in detail:
        return '개발 (일반)'
    elif any(k in detail for k in ['기술개발', '기술직', '기술연구소']):
        return '기술개발 (일반)'
    elif any(k in detail for k in ['제품', '기구개발']):
        return '제품/기구개발 (일반)'
    
    elif 'esg' in detail:
        return '전략/사업개발'
    elif 'eng설계직' in detail or '조선해양' in detail or '엔진 설계' in detail:
        return '설계 - 기계/기구'
    elif '제제연구' in detail or '과학기술' in detail or '연구 개발' in detail:
        return '연구개발 - 화학/바이오'
    elif '배터리 셀' in detail or '배터리 팩' in detail:
        return '연구개발 - 배터리/전지'
    elif '점포개발' in detail:
        return '전략/사업개발'
    elif any(k in detail for k in ['시스템', 'erp개발', '전산직', 'hw', 'smt', '소프트웨어개발']):
        return '시스템개발/IT'
    elif any(k in detail for k in ['타이어설계', '현가시스템', '타이어 구조']):
        return '연구개발 - 차량/자율주행'
    elif '선행기술' in detail:
        return '기술개발 (일반)'
    elif '공정설계' in detail or '배터리 공정' in detail:
        return '연구개발 - 공정'
    elif any(k in detail for k in ['da사업부', 'tech', 'ds 부문', 'h']):
        return '기술연구소/센터'
    elif '교과서' in detail:
        return 'UX/UI/콘텐츠'
    elif '사이버보안' in detail:
        return '연구개발 - 전기/전자'
    elif 'module' in detail:
        return '제품/기구개발 (일반)'
    elif '글로벌' in detail:
        return '전략/사업개발'
    elif '스마트팩토리' in detail:
        return '시스템개발/IT'
    elif 'back-end 개발' in detail:
        return '소프트웨어 개발 - 백엔드'
    elif '회로 설계' in detail:
        return '설계 - 회로'
    elif '생활가전' in detail:
        return '제품개발'


    else:
        return '기타'
```

```python
# 분류 적용
df_it['detail_job'] = df_it['detail_job'].apply(classify_it_job)
df_pro['detail_job'] = df_pro['detail_job'].apply(classify_production_job)
df_re['detail_job'] = df_re['detail_job'].apply(classify_detail_job)
```
