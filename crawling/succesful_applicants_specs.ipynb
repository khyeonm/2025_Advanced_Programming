## 합격자 스펙 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
import re
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time

chrome_options = Options()
chrome_options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=chrome_options)

start_url = "https://linkareer.com/cover-letter/34416?organizationName=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90&page=1&sort=PASSED_AT&tab=all"
driver.get(start_url)

time.sleep(2)

data = []

# 카드 a 태그 리스트 가져오기
cards = driver.find_elements(By.XPATH, '//*[@id="__next"]/div[1]/div[4]/div/div[2]/div/div[1]/div/div[2]/div[1]/div/a')
print(f"카드 개수: {len(cards)}")

# 각 카드의 href 추출
card_urls = []
for idx, card in enumerate(cards, start=1):
    url = card.get_attribute("href")
    print(f"카드 {idx} URL: {url}")
    card_urls.append(url)

# 각 url 접속하여 정보 추출
for idx, url in enumerate(card_urls, start=1):
    driver.get(url)
    time.sleep(2)

    try:
        # ✅ 상세페이지 전용 xpath로 추출
        company_info = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div[4]/div/div[2]/div/div[2]/div[1]/div/div/h1').text
        spec_elements = driver.find_elements(By.XPATH, '//*[@id="__next"]/div[1]/div[4]/div/div[2]/div/div[2]/div[1]/div/div/h3')
        spec_info = spec_elements[0].text if spec_elements else "스펙 없음"

        print(f"\n=== 카드 {idx} ===")
        print("합격 기업 / 직무 / 지원시기:", company_info)
        print("스펙:", spec_info)

        data.append({
            '합격 기업 / 직무 / 지원시기': company_info,
            '스펙': spec_info
        })

    except Exception as e:
        print(f"카드 {idx} 추출 중 오류: {e}")

    # ✅ 리스트 페이지로 돌아가기
    driver.get(start_url)
    time.sleep(2)

# CSV 저장
df = pd.DataFrame(data)
df.to_csv('successful_applicants_specs.csv', index=False, encoding='utf-8-sig')
print("\n✅ CSV 저장 완료!")

driver.quit()
```

## 페이지네이션 추가

```python
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
driver = webdriver.Chrome(options=chrome_options)

base_url = "https://linkareer.com/cover-letter/34416?organizationName=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90&page={}&sort=PASSED_AT&tab=all"

page = 1
all_card_urls = []

while True:
    url = base_url.format(page)
    driver.get(url)
    time.sleep(2)

    # 페이지 카드 가져오기
    cards = driver.find_elements(By.XPATH, '//*[@id="__next"]/div[1]/div[4]/div/div[2]/div/div[1]/div/div[2]/div[1]/div/a')
    if not cards:
        print(f"페이지 {page}: 카드 없음 → 종료")
        break

    print(f"페이지 {page}: 카드 {len(cards)}개 발견")
    for idx, card in enumerate(cards, start=1):
        href = card.get_attribute("href")
        print(f"  카드 {idx}: {href}")
        all_card_urls.append(href)

    page += 1  # 다음 페이지로 이동

print(f"\n총 {len(all_card_urls)}개의 URL 수집 완료!")

# ✅ 각 URL 접속하여 정보 추출
data = []

for idx, url in enumerate(all_card_urls, start=1):
    driver.get(url)
    time.sleep(2)

    try:
        company_info = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div[4]/div/div[2]/div/div[2]/div[1]/div/div/h1').text
        spec_elements = driver.find_elements(By.XPATH, '//*[@id="__next"]/div[1]/div[4]/div/div[2]/div/div[2]/div[1]/div/div/h3')
        spec_info = spec_elements[0].text if spec_elements else "스펙 없음"

        print(f"\n=== {idx}/{len(all_card_urls)} ===")
        print("합격 기업 / 직무 / 지원시기:", company_info)
        print("스펙:", spec_info)

        data.append({
            '합격 기업 / 직무 / 지원시기': company_info,
            '스펙': spec_info
        })

    except Exception as e:
        print(f"URL {idx} 추출 오류: {e}")

# ✅ CSV 저장
df = pd.DataFrame(data)
df.to_csv('successful_applicants_specs.csv', index=False, encoding='utf-8-sig')
print("\n✅ CSV 저장 완료!")

driver.quit()
```

```python
cleaned_data = []

for item in example_data:
    # 1. 합격 기업/직무/지원시기
    parts = [p.strip() for p in item['합격 기업 / 직무 / 지원시기'].split('/')]
    company = parts[0]
    division_job = parts[1].split('_')
    if len(division_job) == 2:
        division, job_title = division_job
    else:
        division, job_title = parts[1], ''
    apply_term = parts[-1]
    
    # 2. 스펙 분해
    spec = item['스펙']
    spec_parts = [p.strip() for p in spec.split('/')]
    university = spec_parts[0]
    major = spec_parts[1] if len(spec_parts) > 1 else ''
    
    # 학점 찾기
    gpa_match = re.search(r'학점\s*([\d.]+)(?:/([\d.]+))?', spec)
    gpa = gpa_match.group(1) if gpa_match else ''
    gpa_scale = gpa_match.group(2) if gpa_match and gpa_match.lastindex == 2 else ''
    
    # 토익
    toeic_match = re.search(r'토익[:]? (\d+)', spec)
    toeic = toeic_match.group(1) if toeic_match else ''
    
    # 오픽
    opic_match = re.search(r'오픽[:]? (\w+)', spec)
    opic = opic_match.group(1) if opic_match else ''
    
    # 토익스피킹
    ts_match = re.search(r'토익스피킹[:]? ([^/]+)', spec)
    toeic_speaking = ts_match.group(1) if ts_match else ''
    
    # 사회생활
    work_match = re.search(r'사회생활 경험[:]? ([^/]+)', spec)
    work_experience = work_match.group(1) if work_match else ''
    
    # 기타 (나머지 자격증 등)
    other = []
    for s in spec_parts:
        if any(keyword in s for keyword in ['자격', '활용능력', 'SQLD', '기사', '운전', '교육', '캡스톤', '논문상', '수상', 'VE', '특허']):
            other.append(s)
    
    cleaned_data.append({
        'company': company,
        'division': division.strip(),
        'job_title': job_title.strip(),
        'apply_term': apply_term,
        'university': university,
        'major': major,
        'gpa': gpa,
        'gpa_scale': gpa_scale,
        'toeic': toeic,
        'opic': opic,
        'toeic_speaking': toeic_speaking,
        'work_experience': work_experience,
        'other_certifications': ', '.join(other)
    })

# 결과 확인
for c in cleaned_data:
    print(c)
```
