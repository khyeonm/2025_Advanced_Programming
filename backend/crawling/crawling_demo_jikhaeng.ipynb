### [기본 테스트용] 열고 닫기만 하는 셀레니움 코드

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# ✅ 크롬 옵션 설정
options = Options()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(options=options)

# ✅ 창 크기 고정 (반응형 대응)
driver.set_window_size(1280, 1024)
wait = WebDriverWait(driver, 20)

# ✅ 사이트 접속
driver.get("https://zighang.com/it")
time.sleep(2)

# ✅ 원하는 직무 설정
job_name = "서버·백엔드"  # 또는 프론트엔드, AI 등

# ✅ 1. 직무 필터 열기
try:
    arrow_xpath = '//*[@id="root"]/main/div[3]/div/div/div/div/div[2]/div/section/button[2]/div/img'
    arrow_button = wait.until(EC.element_to_be_clickable((By.XPATH, arrow_xpath)))
    arrow_button.click()
    print("✅ 직무 필터 열기 성공")
    time.sleep(1)
except Exception as e:
    print(f"❌ 직무 필터 열기 실패: {e}")
    driver.quit()

# ✅ 2. 직무 클릭
try:
    job_button_xpath = f'//button[normalize-space()="{job_name}"]'
    job_button = wait.until(EC.element_to_be_clickable((By.XPATH, job_button_xpath)))
    job_button.click()
    print(f"✅ 직무 '{job_name}' 선택 완료")
    time.sleep(1)
except Exception as e:
    print(f"❌ 직무 선택 실패: {e}")
    driver.quit()

# ✅ 3. 공고 보기 버튼 클릭
try:
    confirm_button = wait.until(EC.presence_of_element_located((
        By.CSS_SELECTOR, 'div.sticky.bottom-0 button.bg-primary'
    )))
    driver.execute_script("arguments[0].scrollIntoView(true);", confirm_button)
    time.sleep(0.3)
    driver.execute_script("arguments[0].click();", confirm_button)
    print("✅ 공고 보기 버튼 클릭 시도 완료")

    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'p.ds-web-title2')))
    print("✅ 공고 리스트 로딩 완료")
    time.sleep(1)

except Exception as e:
    print(f"❌ 공고 보기 버튼 클릭 또는 공고 로딩 실패: {e}")
    driver.quit()

# ✅ 4. 공고 클릭 + 목록 버튼으로 복귀 반복
MAX_CLICKS = 10
original_tab = driver.current_window_handle  # 현재 탭 저장

for i in range(1, MAX_CLICKS + 1):
    try:
        # 공고 제목 기준으로 찾음
        title_xpath = f'(//p[contains(@class, "ds-web-title2")])[{i}]'
        title_elem = wait.until(EC.presence_of_element_located((By.XPATH, title_xpath)))

        # 스크롤 위치 조정
        driver.execute_script("arguments[0].scrollIntoView(true);", title_elem)
        driver.execute_script("window.scrollBy(0, -200);")
        time.sleep(0.3)

        # 상위 a 태그 클릭 (새 탭 열릴 것)
        parent_link = title_elem.find_element(By.XPATH, "./ancestor::a[1]")
        driver.execute_script("arguments[0].click();", parent_link)
        print(f"▶️ [{i}]번째 공고 클릭 → 새 탭 열림 예상")
        time.sleep(2)

        # ✅ 새 탭으로 전환
        all_tabs = driver.window_handles
        new_tab = [tab for tab in all_tabs if tab != original_tab][0]
        driver.switch_to.window(new_tab)
        print(f"🆕 새 탭 전환 완료")

        # 원하는 행동 (ex: 크롤링) 가능
        time.sleep(2)

        # ✅ 새 탭 닫기 & 원래 탭으로 복귀
        driver.close()
        driver.switch_to.window(original_tab)
        print(f"🔙 기존 탭 복귀 완료")
        time.sleep(1)

    except Exception as e:
        print(f"❌ [{i}]번째 공고 실패: {e}")
        continue
```

### 채용공고 긁어오기 추가

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import time
import requests
import pandas as pd

# ✅ 크롬 옵션
options = Options()
options.add_experimental_option("detach", True)
driver = webdriver.Chrome(options=options)
driver.set_window_size(1280, 1024)
wait = WebDriverWait(driver, 20)

# ✅ 사이트 접속
driver.get("https://zighang.com/it")
time.sleep(2)

# ✅ 원하는 직무
job_name = "서버·백엔드"

# ✅ 필터 열기
arrow_xpath = '//*[@id="root"]/main/div[3]/div/div/div/div/div[2]/div/section/button[2]/div/img'
wait.until(EC.element_to_be_clickable((By.XPATH, arrow_xpath))).click()
time.sleep(1)

# ✅ 직무 선택
job_xpath = f'//button[normalize-space()="{job_name}"]'
wait.until(EC.element_to_be_clickable((By.XPATH, job_xpath))).click()
time.sleep(1)

# ✅ 공고 보기 클릭
confirm_button = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.sticky.bottom-0 button.bg-primary')))
driver.execute_script("arguments[0].scrollIntoView(true);", confirm_button)
driver.execute_script("arguments[0].click();", confirm_button)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'p.ds-web-title2')))
time.sleep(1)

# ✅ 공고 탐색
MAX_CLICKS = 10
original_tab = driver.current_window_handle
results = []

for i in range(2, MAX_CLICKS + 1):
    try:
        title_xpath = f'(//p[contains(@class, "ds-web-title2")])[{i}]'
        title_elem = wait.until(EC.presence_of_element_located((By.XPATH, title_xpath)))
        driver.execute_script("arguments[0].scrollIntoView(true);", title_elem)
        driver.execute_script("window.scrollBy(0, -200);")
        parent_link = title_elem.find_element(By.XPATH, "./ancestor::a[1]")
        driver.execute_script("arguments[0].click();", parent_link)
        time.sleep(2)

        # 새 탭으로 이동
        new_tab = [tab for tab in driver.window_handles if tab != original_tab][0]
        driver.switch_to.window(new_tab)
        time.sleep(2)

        # ✅ 데이터 수집
        data = {}
        data["회사명"] = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[1]/div[1]/div[1]/div[1]/div/a').text
        data["경력"] = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[1]/div[1]/div[1]/div[5]/div/section/div[1]/div/div').text
        data["학력"] = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[1]/div[1]/div[1]/div[5]/div/section/div[3]/div/div').text
        data["근무지"] = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[1]/div[1]/div[1]/div[5]/div/section/div[2]/div/div').text
        data["직군"] = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[1]/div[1]/div[1]/div[5]/div/section/div[5]/div/div/div').text

        # ✅ 텍스트 or 이미지 처리
        try:
            prefers = driver.find_element(By.XPATH, '//h2[text()="우대사항"]/following-sibling::p[1]').text
            requires = driver.find_element(By.XPATH, '//h2[text()="자격요건"]/following-sibling::p[1]').text
            data["우대사항"] = prefers
            data["자격요건"] = requires
            data["이미지경로"] = ""
            print(f"✅ [{i}] 텍스트 기반 우대사항/자격요건 크롤링 완료")
        except:
            try:
                img_elem = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[1]/div[1]/div[4]/img')
                img_url = img_elem.get_attribute("src")
                filename = f"{data['회사명']}_공고{i}.png"
                os.makedirs("downloads", exist_ok=True)
                filepath = os.path.join("downloads", filename)
                with open(filepath, "wb") as f:
                    f.write(requests.get(img_url).content)
                data["이미지경로"] = filepath
                data["우대사항"] = ""
                data["자격요건"] = ""
                print(f"🖼️ [{i}] 텍스트 미존재 → 이미지 저장 완료: {filename}")
            except Exception as img_e:
                print(f"❌ [{i}] 이미지 크롤링 실패: {img_e}")

        results.append(data)

        driver.close()
        driver.switch_to.window(original_tab)
        time.sleep(1)

    except Exception as e:
        print(f"❌ [{i}] 공고 처리 실패: {e}")
        driver.switch_to.window(original_tab)
        continue

# ✅ 엑셀 저장
df = pd.DataFrame(results)
df.to_excel("채용공고_크롤링결과.xlsx", index=False)
print("✅ 엑셀 저장 완료")
```
